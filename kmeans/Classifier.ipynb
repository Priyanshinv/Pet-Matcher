```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import pickle
```

```python
#Loading data
df = pd.read_csv("BIG5/data-final.csv", delimiter="\t")
df = df.loc[:,:'OPN10']
df.head()
```

```python
#Remove rows with missing values
df = df.dropna(how='all', axis=0)
df.shape
```

```python
#Reverse Coding
col_list = df.columns
pos_list = col_list[[0,2,4,6,8,11,13,21,23,25,27,28,29,30,32,34,36,38,39,40,42,44,46,48,47,49]]
neg_list = col_list[[1,3,5,7,9,10,12,14,16,18,15,17,19,20,22,24,26,31,33,35,37,41,43,45]]
df[neg_list] = df[neg_list].replace({1:5, 2:4, 3:3, 4:2, 5:1})
```

```python
#Applying kmeans clustering with no of clusters = 5 (as there are 5 personality traits)
classifier = KMeans(n_clusters=5).fit(df)
df['Prediction'] = classifier.labels_
df.head()
```

```python
#Group by clusters
df.groupby('Prediction').mean()
```

```python
ext = col_list[0:10]
est = col_list[10:20]
agr = col_list[20:30]
csn = col_list[30:40]
opn = col_list[40:50]

df_summary = pd.DataFrame()
df_summary['extroversion'] = df[ext].sum(axis=1)/10
df_summary['neurotic'] = df[est].sum(axis=1)/10
df_summary['agreeable'] = df[agr].sum(axis=1)/10
df_summary['conscientious'] = df[csn].sum(axis=1)/10
df_summary['open'] = df[opn].sum(axis=1)/10

df_summary['clusters'] = classifier.labels_
df_summary.groupby('clusters').mean()
```

```python
Observations for each cluster:
Cluster 0: high open, low extro, low neuro
Cluster 1: high agree, high open
Cluster 2: low extro
Cluster 3: high agree
Cluster 4: high agree, low extro, low neuro
```
